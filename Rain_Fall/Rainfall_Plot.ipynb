```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Read the csv file through Pandas and make a dataframe out of it
df = pd.read_csv('Rainfall_Data_Maharastra.csv', low_memory=False)

# to see few contents, we need to print the head of the dataframe
df.head(10)
```

```python
# Now load data for One district 'Ahmednagar' and it's loaded to one dataframe
df_ahmednagar = df[df.District == 'Ahmednagar']
df_ahmednagar
```

```python
# Now Load Year wise data
list_year1 = list(df_ahmednagar[df_ahmednagar.Year == 2004].values)    # Now we have made it to a list
print((list_year1))

```

```python
# Now select our actual data to another list for simplicity
list1 = [] 
for data in (list_year1):
    print(data)
    
```

```python
for data in list_year1[:]:
    for xx in data[3:15]:
        list1.append(float(xx))
        
print(list1)
```

```python
xpos = np.arange(12)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.bar(xpos, list1)
plt.xticks(xpos, months)

plt.savefig('Year_2004_Data_Bar.pdf')
plt.show()
```
